```python
import numpy as np
from scipy.stats import norm
from scipy.stats import poisson
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

# Probability Distribution Fuction
* Probability Density Function (pdf)
* Probability Mass Function (pmf)

## Probability Density Function (pdf) & relation between Cumulative Distribution Function
* It use to check the distribution of the Continuous Random Variable
  * where,
    1) Random Variable
       * It's a process of mapping the output of a random process or experiment to a number, Eg:- Variable(X) = ether '0.0' if head or '1.0' if tail
    2) Contnuious Random Variable
       * floating numbers
       * can't be nagetive

```python
mean = 0
SD = 15
normal_P = np.random.normal(mean,SD,size=1000)
normal_P[:10]
```

```python
normal_P = np.sort(normal_P)

# pdf ploting
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(normal_P, norm.pdf(normal_P, normal_P.mean(), normal_P.std()))
plt.title("Probability Density Function (PDF)")
plt.xlabel("Vaues")
plt.ylabel("Probability Density")

# cdf ploting
plt.subplot(1,2,2)
plt.plot(normal_P, norm.cdf(normal_P, normal_P.mean(), normal_P.std()))
plt.title("Cumulative Distribution Function")
plt.xlabel("Values")
plt.ylabel("Cumulative Probability")
plt.show()
```

## Probability Mass Function (pmf) & relation between Cumulative Distribution Function
* It use to check the distribution of the Discrete Random Variable
  * where,
    1) Random Variable
       * It's a process of mapping the output of a random process or experiment to a number, Eg:- Variable(X) = ether '0' if head or '1' if tail
    2) Discrete Random Variable
       * intiger number
       * can't be float

Taking an example of probabilty of sum of roling two dice

```python
dice = {1,2,3,4,5,6}
sum_of_two_dice = []
for i in dice:
    for j in dice:
        sum_of_two_dice.append(i+j)
unique, frequency = np.unique(sum_of_two_dice, return_counts=True)
unique, frequency
```

```python
plt.figure(figsize=(12,4))

# pdf ploting
plt.subplot(1,2,1)
plt.scatter(unique, frequency/(len(dice)**2))
plt.title("Probability Mass Function (PMF)")
plt.xlabel("Sum of 2 Dice")
plt.ylabel("Probability")

# cdf ploting
plt.subplot(1,2,2)
sum_of_two_dice = np.sort(sum_of_two_dice)
plt.plot(sum_of_two_dice, norm.cdf(sum_of_two_dice, np.mean(sum_of_two_dice), np.std(sum_of_two_dice)))
plt.title("Cumulative Distribution Function (CDF)")
plt.xlabel("Sum of 2 Dice")
plt.ylabel("Cumulative Probability")
plt.show()
```
