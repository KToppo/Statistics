```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
```

# Gaussiam Distribution
* Symmetrical (as in fig.1)

```python
sample_data = np.random.randn(100)
sample_data = np.sort(sample_data)
sample_data
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.histplot(sample_data, kde=True)
plt.title("Histogram")
plt.subplot(1,2,2)
plt.plot(sample_data, norm.pdf(sample_data,sample_data.mean(),sample_data.std()))
plt.title("Normal Distribution curve")
plt.show()
```

# Empirical Rule
* [68-95-99.7]

#### under this rule
- 68% of data lie under (μ-σ to μ+σ) where, μ is mean and σ is standard deviation
- 95% of data lie under (μ-2σ to μ+2σ) where, μ is mean and σ is standard deviation
- 99.7% of data lie under (μ-3σ to μ+3σ) where, μ is mean and σ is standard deviation

### Chaking Empirical Rule in Oue Data-Set

```python
# group_1 = (μ-σ to μ+σ) & group_2 = (μ-2σ to μ+2σ) & group_3 = (μ-3σ to μ+3σ)
group_1 = []
group_2 = []
group_3 = []
mean = sample_data.mean()
SD = sample_data.std()
for i in sample_data:
    if (i>(mean-SD)) & (i<(mean+SD)):
        group_1.append(i)
        group_2.append(i)
        group_3.append(i)
    elif (i>(mean-(2*SD))) & (i<(mean+(2*SD))):
        group_2.append(i)
        group_3.append(i)
    elif (i>(mean-(3*SD))) & (i<(mean+(3*SD))):
        group_3.append(i)
```

```python
group_1_per = len(group_1)/len(sample_data)*100
group_2_per = len(group_2)/len(sample_data)*100
group_3_per = len(group_3)/len(sample_data)*100

group_1_per, group_2_per, group_3_per
```
