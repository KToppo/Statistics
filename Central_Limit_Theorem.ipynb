```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

# Central Limit Theorem
- whether the distribution is gaussion distribution or non-gaussion distribution, if we take a sample where n>=30 and if take all sample mean and where we populated then we get a normal distribution (gaussion model)

```python
mu, sigma = 2, .4
population_data = np.random.lognormal(mu, sigma, 10000)
population_data = np.sort(population_data)
len(population_data)
```

```python
plt.figure(figsize=(12,4))
plt.plot(population_data, norm.pdf(population_data, population_data.mean(), population_data.std()))
plt.title("LogNormal Distribution")
plt.show()
```

```python
# Collecting 100 random samples mean where sample size is 40 each selected from the population_data
sample_mean = []
for i in range(100):
    sample_mean.append(float(np.random.choice(population_data, size=40).mean()))
sample_mean[:10]
```

```python
sample_mean = np.sort(np.array(sample_mean))
sample_mean[:10]
```

```python
plt.figure(figsize=(12,4))
plt.plot(sample_mean, norm.pdf(sample_mean,sample_mean.mean(),sample_mean.std()))
plt.title("Looks like a gaussion distribution?")
plt.show()
```
