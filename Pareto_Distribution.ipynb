```python
import numpy as np
from scipy.stats import norm
from scipy.stats import pareto
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

# Power Law Probability Distribution (Pareto Distribution)
* Used in description of social, quality control, scientific, geophysical, actuarial, and many other types of observable phenomena
* 80-20 Rule
  * Where,
  * 80% of outcomes are due to 20% of causes

```python
alpha1 = 2
xm = 2
pareto_dis = np.random.pareto(alpha1,size=1000)+2
pareto_dis[:10]
```

```python
pareto_dis = np.sort(pareto_dis)
plt.figure(figsize=(12,4))
plt.plot(pareto_dis, pareto.pdf(pareto_dis,alpha1,scale=xm), label=f"Alpha: {alpha1}")
plt.title("Pareto Distribution")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

## 80-20 Rule

```python
pareto_dis.shape
```

```python
plt.figure(figsize=(12,4))
plt.plot(pareto_dis, pareto.pdf(pareto_dis,alpha1,scale=xm),color="black")
plt.fill_between(x=pareto_dis, y1=pareto.pdf(pareto_dis,alpha1,scale=xm),
                 where=(pareto_dis <= (pareto_dis[int(pareto_dis.shape[0]*80/100)])
                       ),
                 label="80% of the Population Data"
                )
plt.fill_between(x=pareto_dis, y1=pareto.pdf(pareto_dis,alpha1,scale=xm),
                 where=(pareto_dis > pareto_dis[int(pareto_dis.shape[0]*80/100)]
                       ),
                 label="Rest 20% of the Population Data"
                )
plt.title("Pareto Distribution")
plt.xlabel("Values")
plt.ylabel("Probability")
plt.legend()
plt.show()
```
